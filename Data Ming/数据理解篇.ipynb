# 查看数据
在数据分析的过程中，第一步都是理解数据。
数据理解包含但不局限于回答以下几个问题：
>1、数据集包含哪些属性？
2、有多少记录？
3、数据的格式是什么？
4、有没有缺失值？
5、这些属性之间有什么联系？


通过观察现有的数据我们发现，至少有两栏存在缺失值，这些缺失值要被填充或丢弃。
还发现许多属性是分类型数据， 分类中使用的大多数算法不能很好地处理分类数据，因此，在数据转换步骤中，我们将需要找到一种方法来将这些数据更改为更适合分类的形式 。
最后，timestamp_first_active属性看起来像一个时间，但是格式不对。
20090609231247应该为2009-06-09 23:12:47
在后续应该进一步处理。

# 更深入挖掘
 现在我们对数据已经有了一个基本的认识，下一步是开始查看数据的结构。
 
 ## Country Destination Values
 查看这个属性每个类别的分布
 通过观察发现，大部分用户的决策被分类了两类
 >不预定或者在US预订
 
 如果进一步把数据细分，我们可以发现：
 >2014年还未预订的用户达到了60%
 
 从建模的角度出发，我们得出以下两点：
 - 类别的分布随着时间改变。考虑到我们最后的结果基于2014年以后的数据，所以要选取时间今的数据来分析。
 - 因为大部分的用户落入两个类别中，为了保证结果的敏感度，要确保模型能够精确预测出其他较少出现的类别。
 
## Account Creation Dates
 
 接下来分析date_account_created这一属性
 
 从图中可以发现数量随时间增长，基本数每个月增长10%，在2014年6月，新创建的用户数量达到125884，比上年增长了132%。
 
 除此之外，我们还可以观察到在训练集中提供的数据都是近两年的。
实际上，如果我们将培训数据限制为从2013年1月起创建的帐户，则我们仍将包含所有数据的70％以上。 这很重要，因为再次参考Airbnb提供的注释，如果我们要使用sessions.csv中的数据，我们将仅限于2014年1月以后的数据。 再次查看这些数字，这意味着，即使sessions.csv数据仅涵盖了11％的时间段（54个月中的6个月），仍然涵盖了30％以上的培训数据，即76,466个用户。

## Age Breakdown
年龄这一属性中存在许多不合常理的值，比如超过1000岁的。通过观察发现，许多人把自己的年龄填成了出生日期。还有一部分人可能处于隐私保护没有填写填写真是年龄。

同时，这一属性还存在大量缺失值。在后续的数据清洗工作中要进行处理。
